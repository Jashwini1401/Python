###1. Discuss string slicing and provide examples.
ANS:- String slicing in Python is a method to extract a specific portion of a string using the syntax string[start:stop:step]. The start index is where the slice begins (inclusive), the stop index is where it ends (exclusive), and step is the interval between characters (optional). This allows efficient string manipulation, such as extracting substrings or reversing a string.

**Example 1:-**

    a = "Ajay"
    print(a[3])
    output:- 'y'
**Example 2:-**

    string 1 = "I am a good student"
    (i)print(string 1[0:4])
    output:-'I am'
    (ii)print(string 1[3: ])
    output:-'m a good student'
**Example 3:-**

    string 2 = "pwskills"
    (i)print(string 2[ :-3])
    output:-'pwski'
    (ii)print(string 2[-3: ])
    output:-'lls'
**Example 4:-**

    string 3 = "hello world!"
    (i)print(string 3[0:5:1])
    output:-'hello'
    (ii)print(string 3[ ::2])
    output:-'hlowrd'
    (iii)print(string 3[ ::-1])
    output:-'!dlrow olleh'


    



###2. Explain the key features of lists in Python.
ANS:- Lists in python can store an ordered collection of items, or elements, of varying types. They are often used to compile multiple items into a single, mutable variable, which is helpful for retrieving items, specifying outputs or performing calculations quickly. Lists are also a type of built-in data structure in Python (along with tuples, sets and dictionaries), which is a specified way of storing and formatting data.

**The key features of Lists in Python:-**
All Python lists include the following features or characteristics:

(i)Lists can contain items of different types at the same time (including strings, integers, floating point numbers and boolean values).

(ii)Lists are mutable and dynamic; list items can be added, removed or changed after the list is defined.

(iii)Lists are ordered; newly added items will be placed at the end of the list.

(iv)Lists use zero-based indexing; every list item has an associated index, and the first item’s index is 0.

(v)Duplicated list items are allowed.

(vi)Lists can be nested within other lists indefinitely.

###3. Describe how to access, modify and delete elements in a list with examples.
ANS:- **(1)Access Values in a Python List:-**Each item in a list has an assigned index value. It’s important to note that Python is a zero-indexed based language. All this means is that the first item in the list starts at index 0 and ascends accordingly.

**Example:-**
    # Define a list
    z = [3, 7, 4, 2]
    # Access the first item of a list at index 0
    (i)print(z[0])
    output:-3
    # Access the first item of a list at index -1
    (ii)print([-1])
    output:-2

  **(2)Modify values in Python List:-**Python lists have different methods that help us modify a list.

  (i)**Index Method:-**The index method returns the first index at which a value occurs.

  **Example:-**
    
    a = [4, 1, 5, 4, 10, 4]
    (i)print(a.index(4))
    output:-0
    (ii)print(a.index(4, 3))
    output:-3
  (ii)**Count Method:-**The count method counts the number of times a value occurs in a list.

  **Example:-**

    list = [4, 1, 5, 4, 10, 4]
    print(list.count(5))
    output:-1
  (iii)**Sort Method:-**The sort method sorts and alters the original list in place.

  **Example:-**

    b = [3, 7, 4, 2]
    b.sort()
    print(b)
    output:-[2,3,4,7]
  (iv)**Append Method:-**The append method adds an element to the end of a list. This happens in place.

  **Example:-**

    c = [7, 4, 3, 2]
    c.append(3)
    print(c)
    output:-[7,4,3,2,3]
  (v)**Extend Method:-**This method extends a list by appending items. The benefit of this is we can add lists together.

  **Example:-**

    list1 = [7, 3, 3]
    list1.extend([4,5])
    print(list1)
    output:-[7,3,3,4,5]
    # Alternatively, the same thing could be accomplished by using the + operator.
    print([1,2] + [3,4])
    output:-[1,2,3,4]
   (vi)**Insert Method:-**The insert method inserts an item before the index we provide.

   **Example:-**

    list2 = [7, 3, 3, 4, 5]
    list2.insert(4, [1, 2])
    print(list2)
    output:-[7,3,3,4,[1,2],5]

**(3)Delete values in Python List:-**Python lists have different methods that help us delete a list.

  (i)**Remove Method:-**The remove method removes the first occurrence of a value in a list.

  **Example:-**

    x = [7, 4, 3, 2, 3]
    x.remove(2)
    print(x)
    output:-[7,4,3,3]
  (ii)**Pop Method:-**The pop method removes an item at the index we provide. This method will also return the item we removed from the list. If we don’t provide an index, it will default to removing the item at the last index.

  **Example:-**

    y = [7, 4, 3, 3]
    print(y.pop(1))
    output:-4
    print(y)
    output:-[7,3,3]

###5. Describe the key features of sets and provide examples of their use.
ANS:-A Set in Python programming is an unordered collection data type that is iterable and has no duplicate elements. While sets are mutable, meaning we can add or remove elements after their creation, the individual elements within the set must be immutable and cannot be changed directly.

Set are represented by { } (values enclosed in curly braces)

The major advantage of using a set, as opposed to a list, is that it has a highly optimized method for checking whether a specific element is contained in the set. This is based on a data structure known as a hash table. Since sets are unordered, we cannot access items using indexes as we do in lists. Set operations like union (combining elements), intersection (finding common elements), and difference (finding elements in one set but not the other) are efficient.

**Example 1 :-**

    Create a Set:
    set = {"apple", "banana", "cherry"}
     print(set)
    output:-{'apple', 'cherry', 'banana'}

    
**Example 2:-** Duplicates values will be Not ignored in sets.Sets cannot have two items with the same value.

    set1 = {"apple", "banana", "cherry", "apple"}
     print(set1)
    output:-{'banana', 'cherry', 'apple'}

**Example 3:-** The values True and 1 are considered the same value in sets, and are treated as duplicates.

    set2 = {"apple", "banana", "cherry", True, 1, 2}
     print(set2)
    output:-{True, 2, 'banana', 'cherry', 'apple'}

**Example 4:-**The values False and 0 are considered the same value in sets, and are treated as duplicates.

    set3 = {"apple", "banana", "cherry", False, True, 0}
     print(set3)
    output:-{False, True, 'cherry', 'apple', 'banana'}


**Example 5:-**To determine how many items a set has,  we use the len() function.

    set4 = {"apple", "banana", "cherry"}
     print(len(set4))
    output:-3


**Example 5:-**Set items can be of any data type.A set with strings, integers and boolean values.

    set5 = {"abc", 34, True, 40, "male"}
      print(set5)
    output:-{'abc', 34, True, 40, 'male'}

**Example 5:-**sets are defined as objects with the data type 'set'.

    set6 = {"apple", "banana", "cherry"}
     print(type(set6))
    output:-<class 'set'>

**Example 6:-**It is also possible to use the set() constructor to make a set.

    set7 = set(("apple", "banana", "cherry"))
     print(set7)
    output:-{'banana', 'apple', 'cherry'}
    # Note: the set list is unordered, so the result will display the items in a random order.

###7.  Describe how to add, modify, and delete items in a dictionary with examples.
ANS:-**Description Of Dictionary:-**

Unordered collections:- Elements are not stored in a specific order.

Unique key-value pairs:- Each key acts as a unique identifier for retrieving an associated value.

Flexible data:- Keys and values can be of various data types (strings, numbers, lists, and even other dictionaries).

**Operations (Basic):-**

Add: Use direct assignment (dictionary_name[key] = value) or .update() method.

Access: Retrieve values using their keys (value = dictionary_name[key]).

Remove: Use del dictionary_name[key], .pop(key), or .popitem().

Check membership: Use the in operator (key in dictionary_name).

**Example:-**

    #student information
    student = {
       "name" : "Bob",
       "age" : 22,
       "courses" : ["Math" , "Chemistry"]
    }
    #Accessing student's age
    age = student["age"]
    print(f"Bob's age: {age}")
    output:- Bob's age = 22

**Analogy:-**

Imagine a fancy key organizer with compartments and labels. Each compartment holds a unique key (like a student ID) and a tag representing the value (like name or class). You can quickly find information (value) by looking up the corresponding key (ID) in the organizer.




###8. Discuss the importance of dictionary keys being immutable and provide examples.
ANS:- To understand the importance of dictionary keys must be immutable,we have to related it to hash table. The hash table implementation of dictionaries uses a hash value calculated from the key value to find the key. If let’s say the key was a mutable object, its value could change, and thus its hash could also change.

As stated above, since whoever changes the key object can’t tell that it was being used as a dictionary key, it can’t move the entry around in the dictionary. Then,

(i)When we try to look up the same object in the dictionary it won’t be found because its hash value is different.

(ii)If we tried to look up the old value it wouldn’t be found either, because the value of the object found in that hash bin would be different.

If you want a dictionary indexed with a list, follow these steps −

(i)Convert the list to a tuple first

(ii)The function tuple(myList) creates a tuple with the same entries as the list myList.

(iii)Tuples are immutable and can therefore be used as dictionary keys.

**Convert List to Tuple:-**

**Example:**

     mylist = ["Jacob", "Harry", "Mark", "Anthony"]

    # Displaying the List

    # Convert List to Tuple
      res = tuple(mylist)
